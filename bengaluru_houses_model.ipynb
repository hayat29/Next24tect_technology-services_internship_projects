```python
def convert_sqrft(string):
    try:
        return float(string)
    except:
        values = string.split('-')
        if len(values) == 2:
            return (float(values[0]) + float(values[1]))/2
    else:
        return None
```

```python
df1 = convert_sqrft('2100 - 2850')
df1
```

```python
#df['total_sqft'] == '2100 - 2850'

df['total_sqft'] = df.total_sqft.apply(convert_sqrft)
df
```

```python
#As the price is mentioned in lakhs, get the price_per_sqrt in a new column
df['price_per_sqrt'] = df['price'] * 100000 / df['total_sqft']
df.head()
```

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
cat_col = df.select_dtypes('object').columns.tolist()
cat_col
```

```python
encoder = OneHotEncoder(min_frequency=10)
encoder.fit(df[cat_col])
```

```python
encoded_col = encoder.get_feature_names_out(cat_col)
encoded_col
```

```python
# Transform the data
encoded = encoder.transform(df[cat_col])
encoded
```

```python
# Convert the result to a DataFrame
encoded_df = pd.DataFrame(encoded.toarray(), columns=encoder.get_feature_names_out(['location']))
encoded_df
```

```python
# Combine the encoded features with the original DataFrame
df_encoded = pd.concat([df.reset_index(drop=True), encoded_df.reset_index(drop=True)], axis=1)
df_encoded
```

```python
df_encoded.drop(columns=['location'], inplace=True)
df_encoded
```

```python
#model building



X = df_encoded.drop(columns = 'price')
y = df_encoded['price']
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
scaler.fit(X)
```
